```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

```python
data = pd.read_csv('/content/mail_data.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
new_data = data.where((pd.notnull(data)),'')
new_data.head()
```

```python
new_data.shape
```

```python
new_data.loc[new_data['Category'] == 'spam', 'Category',] = 0
new_data.loc[new_data['Category'] == 'ham', 'Category',] = 1
```

```python
X = new_data['Message']

Y = new_data['Category']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
feature_extraction.fit(X)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
model = LogisticRegression()
model.fit(X_train_features, Y_train)
```

```python
pred_train = model.predict(X_train_features)
accuracy_train = accuracy_score(Y_train, pred_train)
accuracy_train
```

```python
pred_test = model.predict(X_test_features)
accuracy_test = accuracy_score(Y_test, pred_test)
accuracy_test
```
